```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
df = pd.read_csv("414FINALDIFFSAMPLE.csv")

# Set theme
sns.set_theme(style="whitegrid", context="talk", palette="muted")

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Ranking", y="numberofnotes", palette="plasma_r")
sns.stripplot(data=df, x="Ranking", y="numberofnotes", color="black", size=6, alpha=0.6, jitter=True)

plt.xlabel("Final Ranking")
plt.ylabel("Number of Notes")
plt.title("Note Count by Final Ranking")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("414FINALDIFFSAMPLE.csv")
sns.set_theme(style="white", context="talk", palette="muted")
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x="numberofnotes",
    y="Duration",
    hue = "Duration",
    palette = 'plasma',
    alpha=0.6,
    s=100,
    legend=False
)

#hover over dots to see piece names
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xlabel("Number of Notes")
plt.ylabel("Approximate Duration (seconds)")
plt.title("Note Density per Piece")
plt.tight_layout()
plt.show()
```
