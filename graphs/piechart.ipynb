```python
import matplotlib.pyplot as plt
import numpy as np

competitions = [
    "7th Cliburn Competition", "8th Cliburn Competition", "9th Cliburn Competition",
    "10th Cliburn Competition", "11th Cliburn Competition", "12th Cliburn Competition",
    "13th Cliburn Competition", "14th Cliburn Competition", "15th Cliburn Competition",
    "16th Cliburn Competition", "2015 Tchaikovsky Competition", "2019 Tchaikovsky Competition",
    "2023 Tchaikovsky Competition", "2022 Olga Kern Competition", "2016 Sydney Competition", "2021 Sydney Competition"
]

counts = [36, 31, 32, 36, 33, 22, 25, 29, 26, 26, 32, 25, 24, 22, 21, 26]

cliburn_indices = [i for i, name in enumerate(competitions) if "Cliburn" in name]
other_indices = [i for i in range(len(competitions)) if i not in cliburn_indices]

cliburn_colors = plt.cm.plasma_r(np.linspace(0.4, 0.85, len(cliburn_indices)))
other_colors = plt.cm.viridis(np.linspace(0.4, 0.85, len(other_indices)))

colors = []
#Add functionality for hovering over pie chart slices to display competition names
cliburn_ptr, other_ptr = 0, 0
for i in range(len(competitions)):
    if i in cliburn_indices:
        colors.append(cliburn_colors[cliburn_ptr])
        cliburn_ptr += 1
    else:
        colors.append(other_colors[other_ptr])
        other_ptr += 1


plt.figure(figsize=(10, 7))
plt.pie(counts, labels=None, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Competition Data', fontsize = 24)
plt.axis('equal')
plt.show()


```
