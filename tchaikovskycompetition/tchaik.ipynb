```python
tchaik2019="https://tchaikovskycompetition.com/en/archive-competition/16-competition-2019-god/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
response = requests.get(tchaik2019, headers=headers)
page2= requests.get(tchaik2019, headers=headers)
soup = BeautifulSoup(page2.content, "html.parser")
```

```python
#All Contestant Names here. Dataframe is created here

all_participant_cards19 = soup.find_all("a", {"data-open-modal": True})  # Find all <a> with data-open-modal
#contestantnum=1 #this is for validation
name19_data = []
for card in all_participant_cards19:
    if card["data-open-modal"].startswith("participant-piano-"):  # Check if it starts with "participant-piano-"
        contestantnames2019 = card.find_all("p", class_="member-card__name")
        for name in contestantnames2019:
            #print(name.get_text(strip=True))
            name19_data.append({"Name": name})

df19 = pd.DataFrame(name19_data)
len(name19_data)
#print(df19)
#print(contestantnum) #also for validation

```

```python
country19_data=[]
for card in all_participant_cards19:
    if card["data-open-modal"].startswith("participant-piano-"):  # Check if it starts with "participant-piano-"
        contestantcountry2019 = card.find_all("p", class_="member-card__country")
        for country in contestantcountry2019:
            #print(name.get_text(strip=True))
            country19_data.append({"Country":country})

df19["Country"]=country19_data
```

```python
for card in all_participant_cards19:
    if card.get("data-open-modal", "").startswith("participant-piano-"):
        # Locate the associated modal using `data-modal`
        modal_id = card["data-open-modal"]
        modal = soup.find("div", {"data-modal": modal_id}) #this is how you find it for each separate modal.
        if modal:
            # Now find `accordion__content` within this modal
            test = modal.find("div", class_="accordion__content")

            if test:
                content=test.get_text(separator="CUSTOM", strip=True)
                rounds_list = content.split("CUSTOM")  # Split by "Round I". Had to convert to text first because you can't split a BS4 object.
                print(rounds_list) #this should split it by person.
            else:
                print("No Repetoire")
        else:
            print(f"No modal found for {modal_id}")
```
